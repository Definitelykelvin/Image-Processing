```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
import tkinter as tk
from tkinter import filedialog
from tkinter import *
from PIL import Image
from PIL import ImageTk
import PIL
import pickle

top = tk.Tk()
top.geometry('600x500')
top.title('Rust Detection')
top['bg'] = 'papaya whip'

def upload_image():
    file_path = filedialog.askopenfilename()
    uploaded = Image.open(file_path)
    uploaded.thumbnail(((top.winfo_width()/2.25),(top.winfo_height()/2.25)))
    classify(file_path)
    
def classify(file_path):
    image = Image.open(file_path)
    image = cv2.imread(file_path,0)
    img_copy = cv2.imread(file_path)
    img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
    (ret, mask) =  cv2.threshold(image,80,255,cv2.THRESH_BINARY)
    mask = cv2.bitwise_not(mask)
    se = cv2.getStructuringElement(cv2.MORPH_RECT, (15,15))
    open = cv2.morphologyEx(mask, cv2.MORPH_OPEN, se,iterations=1)
    close = cv2.morphologyEx(open, cv2.MORPH_CLOSE, se,iterations=1)
    edges = cv2.Canny(close,200,300)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for c in contours:
        x, y, w, h = cv2.boundingRect(c)
        if (cv2.contourArea(c)) > 3000:
            heading2.configure(text="Alert: Rusts are detected",font=('Angsana New',16,'bold'))
            heading2.pack() 
            cv2.rectangle(img_copy,(x,y), (x+w,y+h), (255,0,0), 2)
        else :
            heading2.configure(text="Good condition",font=('Angsana New',16,'bold'))
            heading2.pack()  
        
    im = ImageTk.PhotoImage(image=PIL.Image.fromarray(img_copy))
    sign_image.configure(image=im)
    sign_image.image = im


heading = Label(top, text = "Rust Detection", font=('DengXian', 18, 'bold'))
heading['bg'] = 'papaya whip'
heading.pack()

heading2 = Label(top,text="",pady=10)
heading2['bg'] = 'papaya whip'
heading2.pack()

# Result heading
upload = Button(top, text = "Browse image",command= upload_image)
upload.configure(background = 'tomato',height="2", width="20",font=('Arial Black', 14, 'bold'))
upload.place(x=170, y=400)

sign_image = Label(top)
sign_image.pack(expand =True)
sign_image.place(x=150, y=100)
top.mainloop()
```
